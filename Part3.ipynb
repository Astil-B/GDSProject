# Part 3 - Advanced Model

### Taking inspiration from the task-description to use TF-IDF + linear Support Vector Machine.
### This should improve text classification due to altering the weight of exceptionally common words, and performing quite strong on sparse high-dimensional text-data. The same split used in task 2 will be used to keep the comparison fair.

```python
#Loading data and panda-package:
import pandas as pd

df = pd.read_csv("995,000_rows.csv")
```

```python
#Create labels:
df["label"] = df["type"].apply(lambda x: 0 if x == "reliable" else 1)
```

```python
#Splitting dataframe:
from sklearn.model_selection import train_test_split

train_df, temp_df = train_test_split(
    df, test_size=0.2, random_state=69
)

val_df, test_df = train_test_split(
    temp_df, test_size=0.5, random_state=69
)

#Extracting labels
y_train = train_df["label"].values
y_val   = val_df["label"].values
y_test  = test_df["label"].values
```

```python
#Importing packages:
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import f1_score
```

```python
#Using Sklearns TF-IDF vectorizer instead of reinventing the wheel:
vectorizer = TfidfVectorizer(
    max_features=10000,
    ngram_range=(1, 2),
    min_df=2
)
```

```python
#Fit and transform training, test and validation:
x_train_tf = vectorizer.fit_transform(train_df["content"].fillna(""))
x_validation_tf = vectorizer.transform(val_df["content"].fillna(""))
x_test_tf = vectorizer.transform(test_df["content"].fillna(""))
```

```python
#Perform the operation:
model = LinearSVC(C=1.0)
model.fit(x_train_tf, y_train)

val_pred = model.predict(x_validation_tf)
test_pred = model.predict(x_test_tf)
```

```python
#Evaluation, test-score:
val_pred = model.predict(x_validation_tf)
test_pred = model.predict(x_test_tf)
print(f"Validation F1: {(f1_score(y_val, val_pred)):.4f}")
print(f"Test F1: {(f1_score(y_test, test_pred)):.4f}")
```

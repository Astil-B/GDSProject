```python
import pandas as pd
import numpy as np
import sys
import csv
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('punkt_tab')
nltk.download('stopwords')

##task 1 
def load_data(url):
    """Load CSV data from a URL or filepath."""
    csv.field_size_limit(sys.maxsize)
    return pd.read_csv(url, encoding="utf-8", engine="python")


def clean_text(df):
    """Lowercase, strip whitespace, and replace URLs, emails, dates, and numbers."""
    df = df.copy()
    for col in df.select_dtypes(include='object').columns:
        df[col] = df[col].str.lower()
        df[col] = df[col].str.strip()
        df[col] = df[col].str.replace(r"\bhttps?://[^\s]+|www\.[^\s]+", "<URL>", regex=True)
        df[col] = df[col].str.replace(r"[a-zA-Z0-9._%+-]+@[a-zA-Z0-9.-]+\.[A-Za-z]{2,}", "<EMAIL>", regex=True)
        df[col] = df[col].str.replace(r"\b(\d{4}-\d{2}-\d{2}|\d{1,2}[/-]\d{1,2}[/-]\d{2,4})\b", "<DATE>", regex=True)
        df[col] = df[col].str.replace(r"\d+", "<NUM>", regex=True)
    return df


def df_to_string(df):
    """Flatten a df into a single string for tokenization."""
    return "\n".join(
        df.astype(str).apply(lambda row: " ".join(row.dropna().tolist()), axis=1).tolist()
    )


def tokenize(text):
    """Tokenize string and return tokens."""
    return word_tokenize(text)


def remove_stopwords(tokens):
    """Remove stopwords from a list of tokens."""
    stop_words = set(stopwords.words("english"))
    return [word for word in tokens if word.lower() not in stop_words]


def stem_tokens(tokens):
    """Apply Porter stemming to a list of tokens."""
    ps = PorterStemmer()
    return [ps.stem(word) for word in tokens]


def report_vocab_stats(tokens, filtered_tokens, stemmed_tokens):
    """Print vocab size at each processing stage."""
    vocab_size = len(set(tokens))
    vocab_nostop = len(set(filtered_tokens))
    vocab_stem = len(set(stemmed_tokens))

    print(f"Token count:                          {len(tokens)}")
    print(f"Vocabulary size (raw):                {vocab_size}")
    print(f"Vocabulary size (no stopwords):       {vocab_nostop}  "
          f"({((vocab_size - vocab_nostop) / vocab_size) * 100:.3f}% reduction)")
    print(f"Vocabulary size (stemmed):            {vocab_stem}  "
          f"({((vocab_nostop - vocab_stem) / vocab_nostop) * 100:.3f}% reduction)")


def process_corpus(url):
    """Full pipeline: load, clean, tokenize, remove stopwords, stem."""
    df = load_data(url)
    df_clean = clean_text(df)
    text = df_to_string(df_clean)
    tokens = tokenize(text)
    filtered = remove_stopwords(tokens)
    stemmed = stem_tokens(filtered)
    report_vocab_stats(tokens, filtered, stemmed)
    return df_clean, tokens, filtered, stemmed
```

```python
URL = "https://raw.githubusercontent.com/several27/FakeNewsCorpus/master/news_sample.csv"
load_data(URL)
```

```python
df_clean, tokens, filtered_tokens, stemmed_tokens = process_corpus(URL)
```
